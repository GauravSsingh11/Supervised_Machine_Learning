# Import LgisticRegression

First we use train test split method and then using logisticregression.

```python
from sklearn.model_selection import train_test_split
```

We train 70% data and then test on remaining 30% data.

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.3,train_size=0.7,random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
model.score(x_test,y_test)*100
```

# Importing accuracy score classification report and confussion matrix

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
cf=confusion_matrix(y_test,y_pred)
```

```python
cf
```

```python
plt.figure(figsize=(12,5))
sns.heatmap(cf,annot=True,fmt='g',cmap='Blues')
plt.title('Confusion_Matrix',fontsize=20)
plt.show()
```

```python
print(classification_report(y_test,y_pred))
```

```python
accuracy_score(y_test,y_pred)*100
```
